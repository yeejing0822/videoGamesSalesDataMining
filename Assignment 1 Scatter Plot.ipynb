```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('c://VideoGameDatasets1.csv')

df
```

```python
#figure the scatter plot size
plt.figure(figsize=(12,9),dpi = 100)

#create scatter plot
sns.scatterplot(data = df, x = 'Critic_Score', y = 'User_Score', hue='Platform', marker = 'X', alpha=1.0, s= 100)

#label x-axis name and y-axis name
plt.xlabel('Critic Scores for Video Games',fontdict={'fontname':'Comic Sans MS','fontsize':15})
plt.ylabel('User Scores for Video Games',fontdict={'fontname':'Comic Sans MS','fontsize':15})

# give title name
plt.title('Relationship between Critic Score and User Score for the Video Games Platform',fontdict={'fontname':'Comic Sans MS','fontsize':20})

#implement grid in scatter plot
plt.grid()

#save graph as png
plt.savefig('Video Games Scatter Plot.png')

```
