```python
#creating environment
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#read excel file
df = pd.read_csv('D://VideoGameDatasets.csv')
df
```

```python

data = df.groupby('Genre').sum()[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']]

data
```

```python
# define figure and axis of graph
fig, ax = plt.subplots(1, figsize=(12, 10))

# define elements
sales_field = ['NA_Sales','EU_Sales','JP_Sales','Other_Sales']
colours = ['LightBlue', 'Red', 'Purple', 'Black']
legends = ['NA Sales', 'EU Sales', 'JP Sales', 'Others Sales']


# create stacked bar graph
left = len(data) * [0]
for idx, label in enumerate(sales_field):
    plt.barh(data.index, data[label], left = left, color=colours[idx])
    left = left + data[label]

# label title
plt.title('Global Sales for Every Genre of Video Games', loc='left', fontdict={'fontname':'Comic Sans MS','fontsize':14})

# label legend
plt.legend(legends, bbox_to_anchor=([0.45, 1, 0, 0]), ncol=4, frameon=True)

# label x-axis and y-axis name
plt.xlabel('Global Sales',fontdict={'fontname':'Comic Sans MS','fontsize':12})
plt.ylabel('Genre Of Video Games',fontdict={'fontname':'Comic Sans MS','fontsize':12})

# adjust limits & draw grid lines
plt.ylim(-0.5, ax.get_yticks()[-1] + 0.5)
ax.set_axisbelow(True)
ax.xaxis.grid(color='gray', linestyle='dashed')

#save bar graph as png
plt.savefig('Video Games Global Sales Bar.png')

#show graph
plt.show()

```
